```julia
using Distributions
using LinearAlgebra
using Plots
```

```julia
function sim(p,theta, pl = true)
    theta = theta + 1
    x = LinRange(0,1,100)
    y = [theta * i^(theta-1) for i in x]
    x1 = [((rand(1)[1]))^(1/(theta-1)) for i in 1:p]
    y1 = [rand(1)[1] .* theta * i^(theta-1)  for i in x1]

    if pl == true
        plot(x,y, label = "Función de probabilidad de la densidad")
        scatter!(x1,y1, label = "Algoritmo aceptación-rechazo")
    else
        return x1
    end
end
```

```julia
sim(5000,5)
```

Ahora pasamos a verificar que nuestras estimaciones esten correctas con las variables aleatorias que ya simulamos

```julia
media = mean(sim(10000,50,false))
momentos = -(media)/(media-1)
print(momentos)
```

Verificamos que nuestro estimador de maximo verosimil este correcto

```julia
max_ver = 10000/-((sum(log.(sim(10000,50,false)))))
print(max_ver)
```
